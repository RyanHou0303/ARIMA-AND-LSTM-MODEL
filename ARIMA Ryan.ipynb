```python
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_pacf

def pacf_plot(y, lags=30, figsize=(12, 4), style='seaborn'):
    """
    绘制时间序列数据的一次差分图和部分自相关函数（PACF）图。

    参数：
    y : pandas Series
        时间序列数据。
    lags : int, optional
        计算PACF的滞后阶数，默认值为30。
    figsize : tuple, optional
        图形的大小，默认值为(12, 4)。
    style : str, optional
        绘图风格，默认值为'seaborn'。
    """
    # 设置图表风格
    plt.style.use(style)
    
    # 创建一个新的图形对象
    fig, axes = plt.subplots(1, 2, figsize=figsize)
    
    # 绘制时间序列图
    y.plot(ax=axes[0])
    axes[0].set_title('Difference once')
    
    # 绘制部分自相关函数图
    plot_pacf(y, lags=lags, ax=axes[1])
    
    # 自动调整子图的布局，以避免重叠
    plt.tight_layout()
    plt.show()

# 调用函数 pacf_plot，传入差分后的时间序列 diff1
pacf_plot(diff1)
```

```python
from tqdm.notebook import tqdm
from statsmodels.tsa.statespace.sarimax import SARIMAX
def optimise_ARIMA(order_list, exog):
    """
    优化ARIMA模型
    
    :param order_list: 包含(p, d, q)元组的列表，用于尝试不同的ARIMA参数组合
    :param exog: 外生变量

    :return: 包含参数和对应AIC的DataFrame，AIC越低越好
    """
    # 使用列表推导式初始化结果列表
    results = [[order, SARIMAX(exog, order=order).fit(disp=0).aic] 
               for order in tqdm(order_list) 
               if SARIMAX(exog, order=order).fit(disp=0).aic]
    
    # 创建DataFrame
    result_df = pd.DataFrame(results, columns=['(p, d, q)', 'AIC'])
    
    # 按照AIC值排序，AIC越低越好
    result_df = result_df.sort_values(by='AIC').reset_index(drop=True)

    return result_df
```

```python
from itertools import product
# Parameters to iterate through
ps = range(0, 8, 1)
d = 1
qs = range(0, 8, 1)

# Create a list with all possible combination of parameters
parameters = product(ps, qs)
parameters_list = list(parameters)

order_list = []

for each in parameters_list:
    each = list(each)
    each.insert(1, 1)
    each = tuple(each)
    order_list.append(each)
```

```python
result_df = optimise_ARIMA(order_list, exog=data['Adj Close'])
```

```python
start = len(data)  # 开始预测的位置（训练数据的末尾）
end = start + 50   # 预测结束的位置（训练数据的末尾再加上50个时间点）

# 进行预测
predictions = best_model.get_forecast(steps=50)
predicted_mean = predictions.predicted_mean
conf_int = predictions.conf_int()

# 将预测结果的索引调整为正确的时间序列索引
predicted_mean.index = pd.date_range(start=data.index[-1], periods=51, freq='D')[1:]
conf_int.index = pd.date_range(start=data.index[-1], periods=51, freq='D')[1:]
```

```python

# 预测未来一天的股价
forecast_steps = 50
forecast = best_model.get_forecast(steps=forecast_steps)
predicted_mean = forecast.predicted_mean
conf_int = forecast.conf_int()

# 将预测结果的索引调整为正确的时间序列索引
next_day_index = pd.date_range(start=data.index[-1], periods=forecast_steps+1, freq='D')[1:]

predicted_mean.index = next_day_index
conf_int.index = next_day_index

# 打印预测结果
print(f"Next day's predicted stock price: {predicted_mean.iloc[0]}")
print(f"95% confidence interval: {conf_int.iloc[0].values}")

# 绘制原始数据和预测值
plt.figure(figsize=(16, 8))

# 绘制原始数据
plt.plot(data.index, data, label='Observed')

# 绘制预测值
plt.plot(predicted_mean.index, predicted_mean, color='red', marker='o', label='Forecast')

# 绘制置信区间
plt.fill_between(conf_int.index, 
                 conf_int.iloc[:, 0], 
                 conf_int.iloc[:, 1], color='pink', alpha=0.3)

# 添加图例和标题
plt.legend()
plt.title('Forecast vs Observed')
plt.xlabel('Date')
plt.ylabel('Stock Price')

# 显示图形
plt.show()
```
